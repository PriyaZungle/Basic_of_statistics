1. Explain the different types of data(qualitative and quantitative )and provide examples of each. Discuss nominal, oridial, interval and ratio scales  
- Qualitative Data:
  Qualitative data describes qualities or characteristics, often expressed as categories or label
  example :
    - Nominal Data: Categories without a natural order, such as eye color (blue, brown, green), gender (male, female), or types of cars (sedan, SUV, truck.
    - Ordinal Data: Categories with a meaningful order or ranking, but the intervals between them may not be equal, such as education level (high school, bachelor's, master's), customer satisfaction (very dissatisfied, dissatisfied, neutral, satisfied, very satisfied), or letter grades (A, B, C, D,f).
- Quantitative Data:
Quantitative data represents numerical values that can be measured and analyzed.
Examples:
   - Interval Data: Numerical data with equal intervals between values, but an arbitrary zero point (e.g., temperature in Celsius or Fahrenheit).
   - Ratio Data: Numerical data with equal intervals and a true zero point, allowing for meaningful ratios (e.g., height, weight, age, income
2. What are the measures of central tendency, and when should you use.
- Measures of central tendency (mean, median, and mode) describe the "center" of a dataset, each providing a different perspective.
- when to use
     - When you want to use all values in the dataset for a representative value.
     - When you want a measure that is less affected by extreme values.
     - When you have an open-ended distribution.
     - When you want to identify the most common value in a dataset.
3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

- In statistics, dispersion refers to the degree to which a set of data is spread out or clustered around a central value (like the mean).
- Variance measures the average of the squared differences of each data point from the mean. A higher variance indicates that data points are, on average, further away from the mean, meaning the data is more spread out.
- Standard deviation is the square root of the variance. It represents the average distance of data points from the mean in the original units of the data, making it easier to interpret than variance.

4. What is a box plot, and what can it tell you about the distribution of data?
- Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.

5. Discuss the role of random sampling in making inferences about populations.
- Random sampling is crucial for making valid inferences about a population because it helps ensure the sample is representative, minimizing bias and allowing for generalizations to be made about the larger group.

6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
- Skewness, a statistical measure, describes the asymmetry of a data distribution, indicating whether data points lean towards one side of the mean. It has three types: right (positive) skew, left (negative) skew, and zero skew (symmetrical), and its presence can significantly impact data interpretation and analysis.

7. What is the interquartile range (IQR), and how is it used to detect outliers?
- The Interquartile Range (IQR) is the difference between the 75th and 25th percentiles  of a dataset and is used to detect outliers by identifying data points that fall significantly outside the range defined by 1.5 times the IQR.

8. Discuss the conditions under which the binomial distribution is used.

- The binomial distribution is used when you have a fixed number of independent trials, each with only two possible outcomes (success or failure), and a constant probability of success for each trial.
9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).
- The normal distribution is a continuous probability distribution characterized by its bell-shaped curve, and the empirical rule (also known as the 68-95-99.7 rule) provides a quick way to estimate the proportion of data within certain standard deviations of the mean.

11. Explain what a random variable is and differentiate between discrete and continuous random variables.

-A random variable is a variable whose value is a numerical outcome of a random phenomenon, and it can be either discrete (countable values) or continuous (values within a range).
-Discrete Random Variable:
   - A discrete random variable can only take on a finite or countably infinite number of values.
   - The values are typically whole numbers or integers, and they are distinct and separate.
-Continuous Random Variable:
    - A continuous random variable can take on any value within a given range.
    - The values are not distinct, and there are infinitely many possible values between any two values
